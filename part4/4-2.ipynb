```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('./data/__.xlsx')

df.head()

df = df.ffill()
df.head()

mask=(df['']=='')&(df['']!='')
df_seoul=df[mask]
df_seoul=df_seoul.drop([''],axis=1)
df_seoul.rename({'':''},axis=1,inplace=True)
df_seoul.set_index('',inplace=True)
df_seoul

sr_one=df_seoul.loc['']
plt.plot(sr_one.index,sr_one.values,linestyle='--')

plt.title(' ->   ')

plt.xlabel('')
plt.ylabel(' ')

plt.show()
```
