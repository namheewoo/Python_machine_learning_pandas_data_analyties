```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import font_manager,rc
font_path="./data/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)

df = pd.read_excel('./data/__.xlsx')

df = df.ffill()

mask=(df['']=='')&(df['']!='')
df_seoul=df[mask]
df_seoul=df_seoul.drop([''],axis=1)
df_seoul.rename({'':''},axis=1,inplace=True)
df_seoul.set_index('',inplace=True)
df_seoul

sr_one=df_seoul.loc['']

col_years=list(map(str,range(1970,2017)))
df_4=df_seoul.loc[['','','',''],col_years]
df_4=df_4.astype(int)
df_4
df_4=df_4.transpose()

plt.style.use('ggplot')

df_4.plot(kind='area',stacked=True,alpha=0.2,figsize=(20,10))

plt.title(' ->   ',size=30,color='brown',weight='bold')
plt.ylabel('  ',size=20,color='blue')
plt.xlabel('',size=20,color='blue')
plt.legend(loc='best',fontsize=15)

plt.show()
```
