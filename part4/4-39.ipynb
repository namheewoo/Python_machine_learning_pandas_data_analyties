```python
import pandas as pd
import folium

df = pd.read_excel('./data/서울지역_대학교_위치.xlsx', index_col=0)

seoul_map = folium.Map(location=[37.55,126.98], tiles="Cartodb Positron", 
                       zoom_start=12)

for name, lat, lng in zip(df.index, df['위도'], df['경도']):
    folium.CircleMarker([lat, lng],
                        radius=10,             
                        color='brown',        
                        fill=True,
                        fill_color='coral',    
                        fill_opacity=0.7,          
                        popup=name
    ).add_to(seoul_map)

seoul_map
```
