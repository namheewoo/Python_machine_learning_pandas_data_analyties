```python
import pandas as pd

pr_m = pd.period_range(start='2024-01-01',
                       periods=3,
                       end=None,
                       freq='M')
print(pr_m)

pr_h = pd.period_range(start='2024-01-01',
                       periods=3,
                       end=None,
                       freq='h')
print(pr_h)

pr_2d = pd.period_range(start='2024-01-01',
                        end=None,
                        periods=3,
                        freq='2d')
print(pr_2d)
```
