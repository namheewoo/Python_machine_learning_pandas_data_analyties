```python
import pandas as pd

ts_ms = pd.date_range(start='2024-01-01', 
                      end=None,
                      periods=6,
                      freq='MS',
                      tz='Asia/Seoul')
print(ts_ms)
ts_me = pd.date_range(start='2024-01-01', 
                      periods=6,
                      freq='ME',
                      tz='Asia/Seoul')
print(ts_me)
ts_3m = pd.date_range(start='2024-01-01', 
                      periods=6,
                      freq='3MS',
                      tz='Asia/Seoul')
print(ts_3m)                      
```
